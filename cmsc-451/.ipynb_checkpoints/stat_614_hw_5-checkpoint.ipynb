```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv("untitled.txt")
```

```python
plt.hist(df['Time Diff'], bins = 100)
plt.ylabel('number of occurences')
plt.xlabel('time between surgery\nestimate and actual time')
plt.title('Surgery Estimate and Actual Times Histogram')
plt.show()
```

```python
plt.figure(figsize=(20,10))
x = df['Time Diff']
stats.probplot(x, plot=plt)
plt.title('Surgery Estimate and Actual Times Histogram\nProbability Plot', fontsize=20)
plt.show()
```
