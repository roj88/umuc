```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import math
```

```python
df = pd.read_csv("heating.txt", sep='\t')
```

```python
df.head()
```

```python
x = df['index']
y = df['sales']

plt.plot(x, y)
plt.show()
```

```python
for i in df.year.unique():
    x = df['month'].loc[df['year']==i]
    y = df['sales'].loc[df['year']==i]
    plt.plot(x, y, label=i)

plt.legend(loc='upper left')
plt.show()
```

```python
x = df['sales']

plt.hist(x, bins=20)
plt.show()
```

```python
x = df['sales']


np.median(x)
```

```python
for i in df.year.unique():
    y = df['sales'].loc[df['year']==i]
    print("year:{}     mean:{}     median:{}".format(i, round(np.mean(y),1), np.median(y)))
```

```python
for i in df.month.unique():
    y = df['sales'].loc[df['month']==i]
    print("month:{}     mean:{}     median:{}".format(i, round(np.mean(y),1), np.median(y)))
```
